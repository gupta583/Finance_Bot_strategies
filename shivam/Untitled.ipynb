```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('SPY.csv')
# df = pd.read_csv('RawData.xlsx',header=None,error_bad_lines=False)
```

```python
df.describe()
df.columns
```

```python
plt.plot(df['Close'])
plt.ylabel('Price')
plt.xlabel('Time')
```

```python
df['logreturn'] = np.log(df['Close']) - np.log(df['Close'].shift(1))
```

```python
plt.plot(df['logreturn'])
plt.ylabel('Price')
plt.xlabel('Time')
```

```python
import pywt
```
