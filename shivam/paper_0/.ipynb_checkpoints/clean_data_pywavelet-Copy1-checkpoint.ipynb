```python
def waveletSmooth( x, wavelet="haar", level=1, declevel=2):
    coeff = py.wavedec( x, wavelet, mode="per",level=declevel,axis=0 )
    sigma = mad(coeff[-level])
    uthresh = sigma * np.sqrt( 2*np.log( len( x ) ) )
    coeff[1:] = ( py.threshold( i, value=uthresh, mode="soft" ) for i in coeff[1:] )
    y = py.waverec( coeff, wavelet, mode="per",axis=0 )
    f, ax = plt.subplots(figsize=(8,6))
    plt.plot( x, color="b", alpha=0.5)
    plt.plot( y, color="r")
    ax.set_xlim((0,len(y)))
    return y
```

```python
from monthdelta import monthdelta
def interval_split(df):
    dict_dataframes = {}
    split_count = 3
    month_increment = 0
    interval_index = 1
    
#     df = df.set_index(df['date'])
    while split_count > 0:
        front = df['date'][0] 
        front = front + monthdelta(month_increment)
        back = df['date'][0]+monthdelta(30) 
        back = back + monthdelta(month_increment)
        month_increment += 3
        split_count -= 1
        df_interval = pd.DataFrame(df[(df['date'] >= front) & (df['date'] <= back)])
        dict_dataframes[interval_index] = df_interval
        interval_index += 1
        #print(front,back)
    return dict_dataframes
```
